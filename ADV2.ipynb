```python
import numpy as np
import pandas as pd

# plots
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('/content/argentina.csv')
df
```

```python
df.info()
```

### 1. Word Cloud
###A word cloud can be used on the province column to show the distribution of provinces.

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Word Cloud
text = ' '.join(df['province'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

### 2. Box and Whisker Plot
### We'll visualize the distribution of gdp for different provinces.

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.boxplot(x='province', y='gdp', data=df)
plt.xticks(rotation=90)
plt.title('GDP Distribution by Province')
plt.show()
```

### 3. Violin Plot
### We'll plot the distribution of school_dropout rates across provinces.

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='province', y='school_dropout', data=df)
plt.xticks(rotation=90)
plt.title('School Dropout Rate Distribution by Province')
plt.show()
```

### 4. Regression Plot (Linear and Non-linear)
### We'll create a linear regression plot and a polynomial (non-linear) regression plot between gdp and poverty.

```python
# Linear Regression
sns.lmplot(x='gdp', y='poverty', data=df, ci=None)
plt.title('Linear Regression: GDP vs Poverty')
plt.show()
```

```python
#Non-linear Regression (Polynomial)
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

X = df[['gdp']]
y = df['poverty']

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
model = LinearRegression().fit(X_poly, y)

# Predicting values
df['poverty_pred'] = model.predict(X_poly)

plt.figure(figsize=(10, 6))
plt.scatter(df['gdp'], df['poverty'], label='Data', color='blue')
plt.plot(df['gdp'], df['poverty_pred'], color='red', label='Polynomial Fit')
plt.xlabel('GDP')
plt.ylabel('Poverty')
plt.title('Non-linear Regression: GDP vs Poverty')
plt.legend()
plt.show()
```

### 5. 3D Chart
### A 3D scatter plot to visualize gdp, poverty, and doctors_per_cap.

```python
import plotly.express as px

fig = px.scatter_3d(df, x='gdp', y='poverty', z='doctors_per_cap',
                    color='province', size='pop', hover_name='province')
fig.show()
```

### 6. Jitter Plot
### We'll use a jitter plot to show the distribution of birth_mortal.

```python
sns.stripplot(x='province', y='birth_mortal', data=df, jitter=True)
plt.xticks(rotation=90)
plt.title('Birth Mortality by Province')
plt.show()
```

### 7. Line Plot
### We can visualize the trend of illiteracy over provinces.

```python
plt.figure(figsize=(12, 6))
plt.plot(df['province'], df['illiteracy'], marker='o')
plt.xticks(rotation=90)
plt.title('Illiteracy Rate by Province')
plt.show()
```

### 8. Area Plot
### An area plot for the no_healthcare rate across provinces.

```python
plt.figure(figsize=(12, 6))
plt.fill_between(df['province'], df['no_healthcare'])
plt.xticks(rotation=90)
plt.title('No Healthcare Rate by Province')
plt.show()
```

### 9. Donut Chart
### We'll visualize the distribution of population by province.

```python
plt.figure(figsize=(8, 8))
plt.pie(df['pop'], labels=df['province'], autopct='%1.1f%%', wedgeprops={'width': 0.3})
plt.title('Population Distribution by Province')
plt.show()
```

### 10. Treemap
### We'll use a treemap to visualize the distribution of gdp by province.

```python
import plotly.graph_objs as go

fig = go.Figure(go.Treemap(
    labels=df['province'],
    parents=["" for _ in range(len(df))],
    values=df['gdp']
))
fig.show()
```

### 12. Funnel Plot
### We'll visualize the school_dropout rates as a funnel.

```python
fig = px.funnel(df, x='school_dropout', y='province')
fig.show()
```
